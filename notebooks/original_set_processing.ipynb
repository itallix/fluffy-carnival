```python
import pandas as pd
```

### Загружаем оригинальный датасет.

```python
data = pd.read_csv('data/train.tsv', sep='\t', header=None)
data
```

```python
data.columns = ['text', 'labels', 'author']
data
```

```python
data = data [['text', 'labels']]
```

```python
data
```

### Выделяем интересующие нас эмоции

```python
def is_anger(labels):
    if '2' in labels.strip().split(','):
        return 1
    else:
        return 0

data['anger'] = data['labels'].map(is_anger)
data
```

```python
def is_disgust(labels):
    if '11' in labels.strip().split(','):
        return 1
    else:
        return 0

data['disgust'] = data['labels'].map(is_disgust)
data
```

```python
def is_fear(labels):
    if '14' in labels.strip().split(','):
        return 1
    else:
        return 0

data['fear'] = data['labels'].map(is_fear)
data
```

```python
def is_joy(labels):
    if '17' in labels.strip().split(','):
        return 1
    else:
        return 0

data['joy'] = data['labels'].map(is_joy)
data
```

```python
def is_neutral(labels):
    if '20' in labels.strip().split(','):
        return 1
    else:
        return 0

data['neutral'] = data['labels'].map(is_neutral)
data
```

```python
def is_sadness(labels):
    if '26' in labels.strip().split(','):
        return 1
    else:
        return 0

data['sadness'] = data['labels'].map(is_sadness)
data
```

```python
def is_surprise(labels):
    if '27' in labels.strip().split(','):
        return 1
    else:
        return 0

data['surprise'] = data['labels'].map(is_surprise)
data
```

### Отмечаем перекрывающиеся эмоции

```python
def is_multiple(labels):
    return len(labels.strip().split(','))

data['multiple'] = data['labels'].map(is_multiple)
```

```python
data[data['multiple'] > 1]
```

### Суммируем колонки с эмоциями, чтобы потом отфильтровать записи, где нет "наших" эмоций

```python
data['sum'] = data['anger'] + data['disgust'] + data['fear'] + data['joy'] + data['neutral']+data['sadness'] + data['surprise']
```

```python
data[data['sum'] > 0].shape
```

```python
data_nr = data[data['sum'] > 0].copy()
```

```python
data_nr.shape
```

### Датасет уменьшился вдвое. Можно проводить эксперименты с ним, а можно дополнительно отфильтровать сочетания эмоций, где хотя бы одна - не "наша"

```python
data_nr[data_nr['multiple'] > 1].shape
```

```python
data_nr['diff'] = data_nr['multiple'] - data_nr['sum']
```

```python
data_nr[data_nr['diff'] > 0].shape
```

```python
data_r = data_nr.drop(data_nr[data_nr['diff'] > 0].index)
```

```python
data_r.shape
```

### Сохраняем полученные датасеты

```python
data_r.drop(columns=['labels', 'diff', 'sum', 'multiple'], inplace=True)
data_r
```

```python
data_r.to_csv('data/train_r.tsv', sep='\t', index=False)
```
